5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a
heatmap.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generate a synthetic dataset with correlated features
np.random.seed(42)
size = 100  # Number of samples

# Create features with some correlation
x1 = np.random.randn(size)  # Independent feature
x2 = x1 * 0.8 + np.random.randn(size) * 0.2  # Strongly correlated with x1
x3 = np.random.randn(size)  # Independent feature
x4 = x2 * 0.5 + np.random.randn(size) * 0.5  # Moderately correlated with x2
x5 = x1 * -0.6 + np.random.randn(size) * 0.3  # Negatively correlated with x1

# Create a DataFrame
data = pd.DataFrame({
    'Feature 1': x1,
    'Feature 2': x2,
    'Feature 3': x3,
    'Feature 4': x4,
    'Feature 5': x5
})

# Compute the correlation matrix
corr_matrix = data.corr()

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')

# Add title
plt.title('Correlation Matrix Heatmap')

# Display the heatmap
plt.show()
```

3. Using the sales data, generate a heatmap to visualize the variation in sales across different months and
days.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generate the sales data
np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}

# Create a DataFrame
df = pd.DataFrame(data)

# Create a pivot table to organize sales data by Month and Day
pivot_table = df.pivot_table(values='Sales', index='Day', columns='Month', aggfunc='sum', fill_value=0)

# Create a heatmap to visualize sales variation
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_table, annot=True, fmt=".0f", cmap='YlGnBu', linewidths=.5)

# Add title and labels
plt.title('Sales Variation Across Different Months and Days')
plt.xlabel('Month')
plt.ylabel('Day')

# Display the heatmap
plt.show()
```

4. Using the given x and y data, generate a 3D surface plot to visualize the function.

```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go

# Generate x, y, and z data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Flatten the data into a pandas DataFrame
data = {
    'X': x.flatten(),  # Flatten the array
    'Y': y.flatten(),  # Flatten the array
    'Z': z.flatten()   # Flatten the array
}
df = pd.DataFrame(data)

# Create the 3D surface plot using Plotly
fig = go.Figure(go.Surface(z=z, x=x, y=y))

# Customize the layout (optional)
fig.update_layout(
    title='3D Surface Plot of sin(sqrt(x^2 + y^2))',
    scene=dict(
        xaxis_title='X Axis',
        yaxis_title='Y Axis',
        zaxis_title='Z Axis'
    ),
    width=800,
    height=700
)

# Show the plot
fig.show()
```

2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
markers based on the 'sizes' and 'colors' columns.

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import linear_cmap
from bokeh.models import ColumnDataSource
from bokeh.palettes import Viridis256

# Enable Bokeh output in the notebook (optional if using Jupyter)
output_notebook()

# Generate random data
np.random.seed(42)
x = np.random.rand(100) * 10  # x values between 0 and 10
y = np.random.rand(100) * 10  # y values between 0 and 10
sizes = np.random.randint(10, 50, size=100)  # Random sizes for the markers
colors = np.random.randint(0, 256, size=100)  # Random color values

# Create a ColumnDataSource to hold the data
source = ColumnDataSource(data=dict(x=x, y=y, sizes=sizes, colors=colors))

# Create a Bokeh figure
p = figure(title="Bokeh Scatter Plot with Variable Sizes and Colors", x_axis_label='X', y_axis_label='Y')

# Map colors using linear_cmap to scale color values from 0 to 255 across the Viridis256 palette
mapper = linear_cmap(field_name='colors', palette=Viridis256, low=min(colors), high=max(colors))

# Create the scatter plot, using 'sizes' for size and 'colors' for color mapping
p.scatter('x', 'y', size='sizes', color=mapper, source=source, fill_alpha=0.6, legend_label="Random points")

# Show the plot
show(p)
```

4. Create a Bokeh histogram to visualize the distribution of the given data.

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Enable Bokeh output in the notebook (optional if using Jupyter)
output_notebook()

# Generate random data
data_hist = np.random.randn(1000)  # Normal distribution
hist, edges = np.histogram(data_hist, bins=30)  # Create histogram

# Create a figure for the histogram
p = figure(title="Histogram of Random Data", 
           x_axis_label='Value', 
           y_axis_label='Frequency', 
           tools='save', 
           height=400, 
           width=600)

# Add a quad renderer to create the histogram bars
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color='skyblue', line_color='black')

# Show the plot
show(p)
```

5. Create a Bokeh heatmap using the provided dataset.

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import linear_cmap
from bokeh.models import ColorBar
from bokeh.palettes import Viridis256

# Enable Bokeh output in the notebook (optional if using Jupyter)
output_notebook()

# Generate random data for the heatmap
data_heatmap = np.random.rand(10, 10)  # 10x10 random data
x = np.linspace(0, 1, 10)  # x coordinates
y = np.linspace(0, 1, 10)  # y coordinates
xx, yy = np.meshgrid(x, y)  # Create a meshgrid (not used directly for the heatmap)

# Prepare the data for Bokeh
heatmap_data = {'x': xx.flatten(), 'y': yy.flatten(), 'value': data_heatmap.flatten()}

# Create a figure for the heatmap
p = figure(title="Heatmap of Random Data", 
           x_axis_label='X', 
           y_axis_label='Y', 
           width=600, 
           height=600)

# Create a color mapper
mapper = linear_cmap(field_name='value', palette=Viridis256, low=data_heatmap.min(), high=data_heatmap.max())

# Add squares to the heatmap
p.rect(x='x', y='y', width=0.1, height=0.1, source=heatmap_data, fill_color=mapper, line_color='black')

# Add a color bar
color_bar = ColorBar(color_mapper=mapper['transform'], width=8, location=(0, 0))
p.add_layout(color_bar, 'right')

# Show the plot
show(p)
```
